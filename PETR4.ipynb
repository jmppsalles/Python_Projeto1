```python
#importar as bibliotecas
import math
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
#Ler o arquivo em excel e checar se funciona
PETR4 = pd.read_excel(r'Teste_Acoes/COTAHIST_PETR4_2011.xlsx')
print (PETR4)
```

```python
#Plotar grafico do preço de fechamento
plt.figure(figsize=(16,8))
plt.plot(PETR4['Fechamento'])
plt.title('Fechamento PETR4')
plt.xlabel('2011', fontsize=18, labelpad = 10)
plt.ylabel('Fechamento em R$', fontsize=18)
#plt.legend(loc = 'upper left')
plt.grid(b=None, which='major', axis='x')
plt.show()
```

```python
# Criar as médias móveis simples (aritiméticas) de 8, 20 e 100 períodos
mma8 = pd.DataFrame()
mma8['Fechamento'] = PETR4['Fechamento'].rolling(window = 8).mean()
mma20 = pd.DataFrame()
mma20['Fechamento'] = PETR4['Fechamento'].rolling(window = 20).mean()
mma100 = pd.DataFrame()
mma100['Fechamento'] = PETR4['Fechamento'].rolling(window = 100).mean()
```

```python
#Plotar o grafico novamente, mas com as médias móveis
plt.figure(figsize=(17.5,7.5))
plt.plot(PETR4['Fechamento'], label = 'PETR4')
plt.plot(mma8['Fechamento'], label = 'MMA8')
plt.plot(mma20['Fechamento'], label = 'MMA20')
plt.plot(mma100['Fechamento'], label = 'MMA100')
plt.title('Fechamento PETR4')
plt.xlabel('2011')
plt.ylabel('Fechamento em R$')
plt.legend(loc = 'upper right')
plt.show()
```
